```python
one_hot_encoder = OneHotEncoder()
encoded_classes = pd.DataFrame(one_hot_encoder.fit_transform(classes[["Folder Name "]]).toarray())
classes = pd.concat([classes, encoded_classes], axis = 1)
```

```python
# One-hot encoded classes.

classes
```

```python
# A function to convert one-hot encoded classes to bird classes.

def class_decoder(arr):
    for index, item in enumerate(arr):
        if item == 1:
            return classes["Folder Name "].values[index]
```

**iii. In order for all the images to have the same size, zero-pad or resize the images
in your dataset. This can be done using various tools, including OpenCV.**

```python
# Associate the classes with the one-hot encoded classes.

classes_dict = dict()
for value in classes.values:
    classes_dict[value[0]] = value[2:22]
```

```python
directory = "../Data/train-test-val/"
categories = ["train","test", "val"]
    
output = []
    
for category in categories: 
    path = os.path.join(directory, category)
    images = []
    labels = []

    for folder in os.listdir(path): # For each folder (train, test, val),
    
        label = classes_dict[folder]
            
        for file in os.listdir(os.path.join(path, folder)): # read the bird images, convert them into RGB and resize them.
            img_path = os.path.join(os.path.join(path, folder), file)
            
            image = cv2.imread(img_path)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            image = cv2.resize(image, (224,224))
            
            # Append the images and labels to lists.
            
            images.append(image)
            labels.append(label)
            
    images = np.array(images, dtype = "float32")
    labels = np.array(labels, dtype = "int32")
        
    output.append((images,labels))
```

```python
# Retrieve the images and their labels. 

train_images, train_labels = output[0][0], output[0][1]
test_images, test_labels = output[1][0], output[1][1]
val_images, val_labels = output[2][0], output[2][1]
```

**(c) Transfer Learning**

**ii. To perform empirical regularization, crop, randomly zoo, rotate, 
ip, contrast,
and translate images in your training set for image augmentation. You can
use various tools to do this, including OpenCV.**

```python
def image_augmentation(train_images, train_labels):
    boxes = tf.random.uniform((1, 4))
    box_indices = tf.random.uniform((1,), minval=0, maxval = 1, dtype=tf.int32)
    
    for index, image in enumerate(train_images):
        random_crop = random.randint(0, 2) 
        random_zoom = random.randint(0, 2)
        random_translate = random.randint(0, 2)
        random_contrast = random.randint(0, 2)
        random_flip = random.randint(0, 2)
        random_rotate = random.randint(0, 2)
        
        if random_crop == 1:
            augmented_images.append(np.squeeze(tf.image.crop_and_resize(tf.expand_dims(train_images[0],0), boxes, box_indices, (224,224))))
            corresponding_labels.append(train_labels[index])
            
        if random_zoom == 1:
            augmented_images.append(tf.keras.preprocessing.image.random_zoom(image, (0.1,1.0), row_axis = 0, 
                                                                        col_axis = 1, channel_axis = 2, fill_mode = 'nearest',
                                                                        cval = 0.0, interpolation_order = 1))
            corresponding_labels.append(train_labels[index])
            
        if random_translate == 1:
            augmented_images.append(tf.keras.preprocessing.image.random_shift(image, 0.3, 0.3, row_axis = 0,
                                                                        col_axis = 1, channel_axis = 2, fill_mode = 'nearest',
                                                                        cval = 0.0 , interpolation_order = 1))
            corresponding_labels.append(train_labels[index])
            
        if random_contrast == 1:
            augmented_images.append(tf.image.random_contrast(image, 0.1, 0.5, seed=None).numpy())
            corresponding_labels.append(train_labels[index])
            
        if random_flip == 1:
            augmented_images.append(tf.image.random_flip_left_right(image).numpy())
            corresponding_labels.append(train_labels[index])
            
        if random_rotate == 1:
            augmented_images.append(tf.keras.preprocessing.image.random_rotation(image, 45, row_axis = 0,
                                                                        col_axis = 1, channel_axis = 2, fill_mode = 'nearest',
                                                                        cval = 0.0 , interpolation_order = 1))
            corresponding_labels.append(train_labels[index])
        
        if index == 0:
            for augmented_image in augmented_images:
                fig = plt.figure()
                plt.subplot(1,2,1)
                plt.imshow(image/255)
                plt.subplot(1,2,2)
                plt.imshow(augmented_image/255)
                fig.suptitle("Original vs Augmented")
```

```python
augmented_images = []
corresponding_labels = []
image_augmentation(train_images, train_labels)
```

```python
train_images = np.append(train_images, augmented_images, axis = 0)
train_labels = np.append(train_labels, corresponding_labels, axis = 0)
```

```python
# Initialize the EfficientNetB0 model, removes the predictions layer.

efficientnetb0 = EfficientNetB0()
efficientnetb0 = Model(inputs = efficientnetb0.input, outputs = efficientnetb0.get_layer("top_dropout").output)
efficientnetb0.trainable = False
```

```python
# # Freeze all layers except for the last one.

# for layer in efficientnetb0.layers[:-1]:
#     efficientnetb0.get_layer(layer.name).trainable = False
```

**iii. Use ReLU activation functions in the last layer and a softmax layer, along
with batch normalization 4 and a dropout rate of 20% as well as ADAM optimizer. Use multinomial cross entropy loss. You can try any batch size,
but a batch size of 5 seems reasonable.**

```python
# Combine batch normalization, a dropout layer, a dense layer with ReLU activation function, and a softmax layer.

top_dropout = efficientnetb0.get_layer("top_dropout").output
batch_normalization_layer = keras.layers.BatchNormalization()(top_dropout)
dropout_layer = Dropout(rate = 0.2)(batch_normalization_layer)
dense_layer = Dense(20, activation = "ReLU")(dropout_layer)
softmax_layer = Softmax()(dense_layer)
```

```python
# Assemble the final model.

efficientnetb0 = Model(inputs = efficientnetb0.input, outputs = softmax_layer)
```

**iv. Train the networks (EfficientNetB0 and VGG16) for at least 50 epochs
(preferably 100 epochs) and perform early stopping using the validation set.
Keep the network parameters that have the lowest validation error. Plot the
training and validation errors vs. epochs.**

```python
# Train the network with 100 epoch, ADAM optimizer, categorical_crossentropy loss, and early stopping.

early_stopping = EarlyStopping(monitor = 'val_loss', patience = 10)
efficientnetb0.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy", tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
# history = efficientnetb0.fit(datagen.flow(train_images, train_labels, batch_size = 5), epochs = 100, validation_data = (val_images, val_labels), verbose = 2,  callbacks = [early_stopping])
history = efficientnetb0.fit(train_images, train_labels, batch_size = 5, epochs = 100, validation_data = (val_images, val_labels), verbose = 2,  callbacks = [early_stopping])
```

```python
# Plot the errors vs epochs.

def plot_result(item):
    plt.plot(history.history[item], label = item)
    plt.plot(history.history["val_" + item], label = "val_" + item)
    plt.xlabel("Epochs")
    plt.ylabel(item)
    plt.title("Train and Validation {} Over Epochs".format(item), fontsize=14)
    plt.legend()
    plt.grid()
    plt.show()


plot_result("loss")
plot_result("accuracy")
```
