**ii. Randomly select [0.7ni] images from each folder as your training set, [0.15ni]
as validation set, and the rest as your test set, where ni is the number of
images in folder i and [x] is the ceiling of x.**

```python
# As one of the instructors advised, I did not upload the original "images" folder, so please include it if you want to 
# run this cell of code.

folder_path = "../Data/images/"

for folder in os.listdir(folder_path):
    images = [file for file in os.listdir(folder_path + folder)]
    random.shuffle(images)
    folder_length = len(images)
    train_length = math.ceil(folder_length * 0.7)
    val_length = math.ceil(folder_length * 0.15)
    test_length = folder_length - train_length - val_length
        
    old_path = os.path.join("../Data/images/", folder)
    
   
    train_path = os.path.join("../Data/train-test-val/train/", folder)
    test_path = os.path.join("../Data/train-test-val/test", folder)
    val_path = os.path.join("../Data/train-test-val/val", folder)
    
    count = 0
    
    copy_path = train_path
    for image in images:
        
        if count >= train_length:
            copy_path = val_path
        if count >= train_length + val_length:
            copy_path = test_path
        
        if not os.path.exists(copy_path):
            os.makedirs(copy_path)

        shutil.copy(os.path.join(old_path, image), copy_path)
        count += 1
```

```python
# Function to check the number of files in each class.

train_samples = []
test_samples = []
val_samples = []

for folder in os.listdir("../Data/train-test-val/train/"):
    count = 0
    for file in os.listdir("../Data/train-test-val/train/" + folder):
        if os.path.isfile(os.path.join("../Data/train-test-val/train/" + folder , file)):
            count = count + 1
    train_samples.append(count)

for folder in os.listdir("../Data/train-test-val/test/"):
    count = 0
    for file in os.listdir("../Data/train-test-val/test/" + folder):
        if os.path.isfile(os.path.join("../Data/train-test-val/test/" + folder , file)):
            count = count + 1
    test_samples.append(count)
    
for folder in os.listdir("../Data/train-test-val/val/"):
    count = 0
    for file in os.listdir("../Data/train-test-val/val/" + folder):
        if os.path.isfile(os.path.join("../Data/train-test-val/val/" + folder , file)):
            count = count + 1
    val_samples.append(count)
```

```python
# The number of training, validation, and test samples for each class.

pd.DataFrame([train_samples,val_samples,test_samples])
```

**iii. In order for all the images to have the same size, zero-pad or resize the images
in your dataset. This can be done using various tools, including OpenCV.**

```python
# Associate the classes with the one-hot encoded classes.

birds_dict = dict()
classes_dict = dict()
for value in classes.values:
    birds_dict[value[0]] = value[2:22]
    classes_dict[value[1]] =  value[2:22]
```

```python
directory = "../Data/train-test-val/"
categories = ["train","test", "val"]
    
output = []
    
for category in categories: 
    path = os.path.join(directory, category)
    images = []
    labels = []
    
    # For each folder (train, test, val):
    
    for folder in os.listdir(path): 
    
        label = birds_dict[folder]
        
        for file in os.listdir(os.path.join(path, folder)): 
            img_path = os.path.join(os.path.join(path, folder), file)
            
            # Read the images and resize them to (224,224),
            
            image = cv2.imread(img_path)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            image = cv2.resize(image, (224,224))
            
            # Append the images and labels to lists.
            
            images.append(image)
            labels.append(label)
            
    images = np.array(images, dtype = "float32")
    labels = np.array(labels, dtype = "int32")
        
    output.append((images,labels))
```

```python
# Retrieve the images and their labels. 

train_images, train_labels = output[0][0], output[0][1]
test_images, test_labels = output[1][0], output[1][1]
val_images, val_labels = output[2][0], output[2][1]
```

**(c) Transfer Learning**

**ii. To perform empirical regularization, crop, randomly zoo, rotate, 
ip, contrast,
and translate images in your training set for image augmentation. You can
use various tools to do this, including OpenCV.**

```python
# A function to randomly augment images.
# Each augmentation method gets a 1/3 chance of being activated, therefore, for each image, usually, 2 more samples are created.
# The number of samples created can vary randomly.

def image_augmentation(train_images, train_labels):
    boxes = tf.random.uniform((1, 4))
    box_indices = tf.random.uniform((1,), minval=0, maxval = 1, dtype=tf.int32)
    
    for index, image in enumerate(train_images):
        random_crop = random.randint(0, 2) 
        random_zoom = random.randint(0, 2)
        random_translate = random.randint(0, 2)
        random_contrast = random.randint(0, 2)
        random_flip = random.randint(0, 2)
        random_rotate = random.randint(0, 2)
        
        if random_crop == 1:
            augmented_images.append(np.squeeze(tf.image.crop_and_resize(tf.expand_dims(train_images[0],0), boxes, box_indices, (224,224))))
            corresponding_labels.append(train_labels[index])
            
        if random_zoom == 1:
            augmented_images.append(tf.keras.preprocessing.image.random_zoom(image, (0.1,1.0), row_axis = 0, 
                                                                        col_axis = 1, channel_axis = 2, fill_mode = 'nearest',
                                                                        cval = 0.0, interpolation_order = 1))
            corresponding_labels.append(train_labels[index])
            
        if random_translate == 1:
            augmented_images.append(tf.keras.preprocessing.image.random_shift(image, 0.3, 0.3, row_axis = 0,
                                                                        col_axis = 1, channel_axis = 2, fill_mode = 'nearest',
                                                                        cval = 0.0 , interpolation_order = 1))
            corresponding_labels.append(train_labels[index])
            
        if random_contrast == 1:
            augmented_images.append(tf.image.random_contrast(image, 0.1, 0.5, seed=None).numpy())
            corresponding_labels.append(train_labels[index])
            
        if random_flip == 1:
            augmented_images.append(tf.image.random_flip_left_right(image).numpy())
            corresponding_labels.append(train_labels[index])
            
        if random_rotate == 1:
            augmented_images.append(tf.keras.preprocessing.image.random_rotation(image, 45, row_axis = 0,
                                                                        col_axis = 1, channel_axis = 2, fill_mode = 'nearest',
                                                                        cval = 0.0 , interpolation_order = 1))
            corresponding_labels.append(train_labels[index])
        
        if index == 0:
            for augmented_image in augmented_images:
                fig = plt.figure()
                plt.subplot(1,2,1)
                plt.imshow(image/255)
                plt.subplot(1,2,2)
                plt.imshow(augmented_image/255)
                fig.suptitle("Original vs Augmented")
```

```python
# Create the augmented images - labels and plot the first image's augmented version(s).

augmented_images = []
corresponding_labels = []
image_augmentation(train_images, train_labels)
```

```python
# Append the augmented images to the original images.

train_images = np.append(train_images, augmented_images, axis = 0)
train_labels = np.append(train_labels, corresponding_labels, axis = 0)
```

## EfficientNetB0

```python
# Initialize the EfficientNetB0 model, remove the last layer.

efficientnetb0 = EfficientNetB0()
efficientnetb0 = Model(inputs = efficientnetb0.input, outputs = efficientnetb0.get_layer("top_dropout").output)
efficientnetb0.trainable = False
```

**iii. Use ReLU activation functions in the last layer and a softmax layer, along
with batch normalization 4 and a dropout rate of 20% as well as ADAM optimizer. Use multinomial cross entropy loss. You can try any batch size,
but a batch size of 5 seems reasonable.**

```python
# Combine batch normalization, a dropout layer, a dense layer with ReLU activation function, and a softmax layer.

top_dropout = efficientnetb0.get_layer("top_dropout").output
batch_normalization_layer = keras.layers.BatchNormalization()(top_dropout)
dropout_layer = Dropout(rate = 0.2)(batch_normalization_layer)
dense_layer = Dense(20, activation = "ReLU")(dropout_layer)
softmax_layer = Softmax()(dense_layer)
```

```python
# Assemble the final model.

efficientnetb0 = Model(inputs = efficientnetb0.input, outputs = softmax_layer)
```

**iv. Train the networks (EfficientNetB0 and VGG16) for at least 50 epochs
(preferably 100 epochs) and perform early stopping using the validation set.
Keep the network parameters that have the lowest validation error. Plot the
training and validation errors vs. epochs.**

```python
# Train the network with 100 epoch, ADAM optimizer, categorical_crossentropy loss, early stopping and save the best weights.

efficientnetb0_checkpoint_path = '../weights/EfficientNetB0/'

efficientnetb0_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(filepath = efficientnetb0_checkpoint_path, 
                                                               save_weights_only = True, monitor = 'val_accuracy',
                                                               save_best_only=True)

early_stopping = EarlyStopping(monitor = 'val_loss', patience = 10, restore_best_weights = True)
efficientnetb0.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy", tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
history = efficientnetb0.fit(train_images, train_labels, batch_size = 5, epochs = 100, validation_data = (val_images, val_labels), verbose = 2,  callbacks = [early_stopping, efficientnetb0_checkpoint_callback])
```

```python
efficientnetb0_training_errors = [1 - accuracy for accuracy in history.history["accuracy"]]
efficientnetb0_validation_errors = [1 - accuracy for accuracy in history.history["val_accuracy"]]
```

```python
# Plot the errors vs epochs.

plt.plot(efficientnetb0_training_errors, label = "training_error")
plt.plot(efficientnetb0_validation_errors, label = "validation_error")
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Training and validation errors vs Epochs")
plt.legend()
plt.grid()
plt.show()
```

```python
# Test the model on the test set.

efficientnetb0.evaluate(test_images, test_labels)
```

**v. Report Precision, Recall, and F1 score for your model. Remember that this
is a multi-class classification problem.**

```python
# Get the softmax layer's output.

predictions = efficientnetb0.predict(test_images)
```

```python
target_classes = classes["Folder Name "].values
```

```python
# Convert the probabilities into one-hot encoded classes.

predicted_labels = []
for item in np.argmax(predictions, axis = 1):
    predicted_labels.append(list(classes_dict[item]))
```

```python
# Precision, Recall, and F1-score for all 20 classes.

print(classification_report(test_labels, predicted_labels, target_names = target_classes))
```

## VGG16

```python
# Initialize the VGG16 model, remove the last layer.

vgg16 = VGG16()
vgg16 = Model(inputs = vgg16.input, outputs = vgg16.get_layer("fc2").output)
vgg16.trainable = False
```

**iii. Use ReLU activation functions in the last layer and a softmax layer, along
with batch normalization 4 and a dropout rate of 20% as well as ADAM optimizer. Use multinomial cross entropy loss. You can try any batch size,
but a batch size of 5 seems reasonable.**

```python
# Combine batch normalization, a dropout layer, a dense layer with ReLU activation function, and a softmax layer.

fc2 = vgg16.get_layer("fc2").output
batch_normalization_layer = keras.layers.BatchNormalization()(fc2)
dropout_layer = Dropout(rate = 0.2)(batch_normalization_layer)
dense_layer = Dense(20, activation = "ReLU")(dropout_layer)
softmax_layer = Softmax()(dense_layer)
```

```python
# Assemble the final model.

vgg16 = Model(inputs = vgg16.input, outputs = softmax_layer)
```

**iv. Train the networks (EfficientNetB0 and VGG16) for at least 50 epochs
(preferably 100 epochs) and perform early stopping using the validation set.
Keep the network parameters that have the lowest validation error. Plot the
training and validation errors vs. epochs.**

```python
# Preprocess input for VGG16.

train_images_vgg16 = preprocess_input(train_images)
val_images_vgg16 = preprocess_input(val_images)
test_images_vgg16 = preprocess_input(test_images)
```

```python
# Train the network with 100 epoch, ADAM optimizer, categorical_crossentropy loss, early stopping and save the best weights.
vgg16_checkpoint_path = '../weights/VGG16/'

vgg16_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(filepath = vgg16_checkpoint_path, 
                                                               save_weights_only = True, monitor = 'val_accuracy',
                                                               save_best_only=True)

early_stopping = EarlyStopping(monitor = 'val_loss', patience = 10, restore_best_weights = True)
vgg16.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy", tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
history = vgg16.fit(train_images_vgg16, train_labels, batch_size = 5, epochs = 100, validation_data = (val_images_vgg16, val_labels), verbose = 2,  callbacks = [early_stopping, vgg16_checkpoint_callback])
```

```python
vgg16_training_errors = [1 - accuracy for accuracy in history.history["accuracy"]]
vgg16_validation_errors = [1 - accuracy for accuracy in history.history["val_accuracy"]]
```

```python
# Plot the errors vs epochs.

plt.plot(vgg16_training_errors, label = "training_error")
plt.plot(vgg16_validation_errors, label = "validation_error")
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Training and validation errors vs Epochs")
plt.legend()
plt.grid()
plt.show()
```

```python
# Test the model on the test set.

vgg16.evaluate(test_images_vgg16, test_labels)
```
