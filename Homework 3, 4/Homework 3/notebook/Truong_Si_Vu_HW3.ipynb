**ii. Extract the time-domain features minimum, maximum, mean, median, standard deviation, first quartile, and third quartile for all of the 6 time series in each instance. You are free to normalize/standardize features or use them directly.**

```python
features = list(bending1_dataset1.columns)[1:]
features
```

```python
# Function to extract time-domain features from the datasets.

def time_domain_feature_extraction(dataframe):
    row_results = []
    for feature in features:
        minn = dataframe[feature].min()
        maxx = dataframe[feature].max()
        mean = dataframe[feature].mean()
        median = dataframe[feature].median()
        std = dataframe[feature].std()
        first_quant = list(dataframe[feature].quantile([0.25]))[0]
        third_quant = list(dataframe[feature].quantile([0.75]))[0]
        
        row_results.extend([minn, maxx, mean, median, std, first_quant, third_quant])
        
    return row_results
```

```python
# Extract all the time-domain features.

time_domain_features_list = []
for dataset in all_datasets:
    for instance in dataset:
        time_domain_features_list.append(time_domain_feature_extraction(instance))
```

**iii. Estimate the standard deviation of each of the time-domain features you extracted from the data. Then, use Python's bootstrapped or any other method to build a 90% bootsrap confidence interval for the standard deviation of each feature.**

```python
stds = list(time_domain_features[columns_tdf].describe().loc["std"])
```

```python
# Standard Deviation.

stds_df = pd.DataFrame(zip(columns_tdf,stds), columns = ["feature", "standard deviation"])
stds_df
```

```python
# Confidence Intervals.

cis = []

for feature in columns_tdf:
    new_feature = (list(time_domain_features[feature]),)
    confidence_interval = bootstrap(new_feature, np.std, confidence_level=0.9 ,method = "basic")
    cis.append((confidence_interval.confidence_interval[0], confidence_interval.confidence_interval[1]))
```

```python
cis_df = pd.DataFrame(zip(columns_tdf, cis), columns = ["feature", "confidence interval"])
cis_df
```
